```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Check the unique values

unique_values = []
for i in df.columns:
    unique_count = df[i].value_counts().count()
    unique_values.append([i,unique_count])
```

```python
print(unique_values)
```

```python
matrix = df.corr().round(2)
sns.set(rc={"figure.figsize":(20,10)})
sns.heatmap(matrix, annot=True, cmap="YlOrBr")
plt.show()
```
